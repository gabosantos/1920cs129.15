## Preprocessing the Dataframe
### A. Dealing with Missing Values
We will impute the missing data with the mean of that column. You can also use other centrality measures.  We will use _fillna()_ to impute values to the missing values.  __[*fillna()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.fillna.html)__ accepts a parameter of the value to be imputed. For example:

```python
df["Age"] = df["Age"].fillna(value = df["Age"].mean())
```

Here, we called the "Age" column by using square brackets beside our dataframe variable _df_.  At the same time, we can replace a column in our dataframe by calling the column that we want to replace.  In our example above, **df["Age"]** is called and assigned back to column **df["Age"]**.

To get the mean of the Age column, we use the __[*mean()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.mean.html)__ method; hence, we have have **df["Age"].mean()** as the parameter value to _fillna()_.  After imputing the missing values using the parameter value, it creates a new dataframe column and assigns it back to the Age column.

Let's confirm if we still have missing values in our Age column using another method, __[*isnull()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.isnull.html)__.__[*sum()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.sum.html)__.

```python
df["Age"].isnull().sum()
```

This confirms that there are no more missing values in our *Age* column. Now let's investigate the second column with missing values, the *Embarked* column. Similarly, we need to decide what value to impute for the missing values in this column.  But since this is a categorical variable, we can do two things:
1. Assume the missing values have the most probable value
2. Create a new value for missing values

In this notebook, we will use the first method. To get the most probable value also means getting the mode of that column.  We run the script below:

```python
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace = True)
```

Let's confirm once again if there are still missing values left in the *Embarked* column.

```python
df["Embarked"].isnull().sum()
```

This confirms that there are no more missing values in our *Embarked* column.
### B. Performing One-Hot Encoding on Categorical Values
Before we perform one-hot encoding on nominal values, let's look again at our dataframe.

```python
df.head()
```

Both *Name* and *Ticket* columns are unique identifiers to each passenger in Titanic.  But we also have a *PassengerId* as a numeric unique identifier.  Since our objective is to have a dataframe filled with numeric values, we can drop *Name* and *Ticket* columns and just leave the *PassengerId*.  

Another nominal column is the *Cabin* column.  Currently, not all passengers are assigned to a cabin and we don't have an idea on the order or hierarchy of the cabins.  It is safe to assume for now that we don't need this column as well.

To drop the three columns, we will use __[*drop()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html)__.

```python
df.drop(["Name", "Ticket", "Cabin"], axis = 1, inplace = True)
```

The first parameter to _drop()_ is the list of columns that we want to drop, enclosed in square brackets. The second parameter dictates that we are dropping a column (axis=1).  The third parameter states that we assign the new dataframe without the dropped columns back to the original dataframe variable.

If you run _df.head()_ again, we don't see the dropped columns anymore.

```python
df.head()
```

We can now proceed with performing one-hot encoding for the 2 remaining nominal variables.  To do that, we will use __[*get_dummies()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.get_dummies.html)__ method.

```python
df = pd.get_dummies(df)
```

Let's take a look at our updated dataframe.

```python
df.head()
```

Now, we can confirm that all the values in our dataframe are numeric.
### C. Detecting Outliers
Let's make the boxplot visualization again using our updated dataframe.

```python
df.boxplot()
```

Here we can see that among the columns with suspected outliers, only *Age* and *Fare* columns have the most spread and evident outliers.  For now, we will ignore the other columns and just deal with the columns *Age* and *Fare*.

In dealing with these outliers, we will just drop the rows containing them. Let's identify first the upper and lower bounds for both columns. The upper bound is defined as $75\% Percentile + 1.5*IQR$, while the lower bound is defined as $25\% Percentile - 1.5*IQR$.  We can get the quartiles using __[*quantile()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.quantile.html)__. 

```python
IQR_Fare = df["Fare"].quantile(0.75) - df["Fare"].quantile(0.25)
UpperBound_Fare = df["Fare"].quantile(0.75) + 1.5*IQR_Fare
LowerBound_Fare = 0 # There is no negative fare so we clip it to zero.

print("IQR:", IQR_Fare, ", Lower Bound:", LowerBound_Fare, ", Upper Bound:", UpperBound_Fare)
```

Using these bounds, we can select which rows to include and which rows to drop.

```python
df = df[df["Fare"] <= UpperBound_Fare][df["Fare"] >= LowerBound_Fare]
df
```

The condition on the first square brackets selects all rows below the upper bound, including the upper bound. The condition on the second square brackets then selects all rows above the lower bound, including the lower bound, on the resulting dataframe from the first condition. As a result of dropping the outliers on the *Fare* column, we are left with 775 rows.

Let's do the same thing with the *Age* column.

```python
IQR_Age = df["Age"].quantile(0.75) - df["Age"].quantile(0.25)
UpperBound_Age = df["Age"].quantile(0.75) + 1.5*IQR_Age
LowerBound_Age = df["Age"].quantile(0.25) - 1.5*IQR_Age

print("IQR:", IQR_Age, ", Lower Bound:", LowerBound_Age, ", Upper Bound:", UpperBound_Age)
```

```python
df = df[df["Age"] <= UpperBound_Age][(df["Age"] >= LowerBound_Age)]
df
```

The number of rows have decreased further to 708 rows due to removing the outliers in the *Age* column.  

Let's review the boxplot visualization again, now with the filtered dataframe.

```python
df.boxplot(column=["Age", "Fare"])
```

What we are seeing here is that after removing the outliers from the original dataframe, we have arrived at a totally new distribution.  We can transform the original distribution by taking the square root or log of data values, instead of removing them completely. For now, we will end here and assume that we have removed the outliers from the original distribution.
### D. Normalizing the data
We have seen that the numeric values are at different scales. Let's work on normalizing the numeric columns. One of the columns, the *PassengerId*, does not need to be normalized since it's a unique identifier. So, we will use it as the new index now using __[*set_index()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.set_index.html)__.

```python
df.set_index("PassengerId", inplace=True)
df
```

Now we can proceed with normalizing our data.  We will use min-max feature scaling to give our columns a hard range of [0,1].

First, we will use a **for** loop to iterate over the columns of our dataframe. Then, each column will be replaced with the new and normalized column.

```python
for column in df.columns:
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
```

Let's print a summary of our dataframe using *describe()*.

```python
df.describe()
```

Now, we can verify that all columns are now within [0,1].  This final dataset is now ready for model creation. Let's export this dataframe to a new csv file using __[*to_csv()*](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.to_csv.html)__.

```python
df.to_csv("../output/titanic_cleaned.csv")
```
